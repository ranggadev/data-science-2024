#Q2. Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?

```python
# ============================================================
# Q2. Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?
# ============================================================
missing_info = pd.DataFrame({
    'Kolom': df.columns,
    'Jumlah_Missing': df.isnull().sum(),
    'Proporsi_%': (df.isnull().sum() / len(df) * 100).round(2)
})

missing_info = missing_info[missing_info['Jumlah_Missing'] > 0].sort_values('Jumlah_Missing', ascending=False)

if len(missing_info) > 0:
    print("\nKolom dengan missing values:")
    print(missing_info.to_string(index=False))
else:
    print("\nTidak ada missing values")
```

Penjelasan:

Kolom dengan missing value adalah "Age, Cabin, dan Embarked".

- Kolom "Age" missing valuenya 177 dan propostinya 19.87%
- Kolom "Cabin" missing valuenya 687 dan propostinya 77.10%
- kolom "Embarked" missing valuenya 2 dan propostinya 0.22%

#Q3. Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.

```python
# ============================================================
# Q3. Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.
# ============================================================
# Metode 1: Menggunakan Pandas
persen_selamat_pd = (df['Survived'].sum() / len(df)) * 100
jumlah_selamat_pd = df['Survived'].sum()
jumlah_tidak_selamat = len(df) - jumlah_selamat_pd

# Metode 2: Menggunakan NumPy
survived_array = np.array(df['Survived'])
persen_selamat_np = (np.sum(survived_array) / len(survived_array)) * 100

print(f"\nJumlah penumpang selamat: {int(jumlah_selamat_pd)}")
print(f"Jumlah penumpang tidak selamat: {int(jumlah_tidak_selamat)}")
print(f"Persentase selamat (Pandas): {persen_selamat_pd:.2f}%")
print(f"Persentase selamat (NumPy): {persen_selamat_np:.2f}%")

# Visualisasi tabel
survival_count = df['Survived'].value_counts()
print(f"\nRingkasan:")
print(f"  Tidak selamat: {survival_count.get(0, 0)} penumpang")
print(f"  Selamat: {survival_count.get(1, 0)} penumpang")
```

Penjelasan:

- Penumpang selamat sebanyak 38.38% (342 dari 891 orang)
- Sedangkan yang tidak selamat sebanyak 549 orang

#Q4. Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
# ============================================================
# Q4. Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?
# ============================================================
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
jumlah = len(wanita_kelas1)
rata_usia = wanita_kelas1['Age'].mean()

print(f"\nJumlah penumpang wanita kelas 1: {jumlah}")
print(f"Rata-rata usia: {rata_usia:.2f} tahun")
print(f"\nDetail penumpang wanita kelas 1:")
print(wanita_kelas1[['Name', 'Age', 'Sex', 'Pclass', 'Survived']].to_string(index=False))
```

Penjelasan:
- Jumlah penumpang wanita kelas 1: 94
- Rata-rata usia: 34.61 tahun

#Q5. Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

```python
# ============================================================
# Q5. Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?
# ============================================================
tingkat_selamat = df.groupby('Pclass')['Survived'].agg([
    ('Jumlah_Penumpang', 'count'),
    ('Jumlah_Selamat', 'sum'),
    ('Tingkat_Selamat_%', lambda x: (x.sum() / len(x)) * 100)
]).round(2)

print("\n")
print(tingkat_selamat)

kelas_tertinggi = df.groupby('Pclass')['Survived'].mean().idxmax()
nilai_tertinggi = df.groupby('Pclass')['Survived'].mean().max()

print(f"\nKelas dengan tingkat keselamatan tertinggi: Kelas {int(kelas_tertinggi)}")
print(f"Tingkat keselamatan: {nilai_tertinggi * 100:.2f}%")

# Analisis tambahan: Perbandingan per kelas dan gender
print("\n--- Analisis Tambahan: Tingkat Keselamatan per Kelas dan Gender ---")
analisis_detail = df.groupby(['Pclass', 'Sex'])['Survived'].agg([
    ('Jumlah', 'count'),
    ('Selamat', 'sum'),
    ('Tingkat_%', lambda x: (x.sum() / len(x)) * 100)
]).round(2)
print(analisis_detail)
```

Penjelasan:
- Kelas 1: 62.96%
- Kelas 2: 47.28%
- Kelas 3: 24.24%

Jadi tingkat keselamatan yang paling tinggi adalah **Kelas 1**

#Q6. Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.

```python
# ============================================================
# Q6.Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.
# ============================================================
# PENGARUH USIA TERHADAP KESELAMATAN

# Membuat kategori usia
df['Kelompok_Usia'] = pd.cut(df['Age'],
                               bins=[0, 12, 18, 35, 50, 100],
                               labels=['Anak (0-12)', 'Remaja (13-18)',
                                      'Dewasa (19-35)', 'Dewasa Tua (36-50)',
                                      'Lansia (>50)'])

analisis_usia = df.groupby('Kelompok_Usia', observed=True)['Survived'].agg([
    ('Jumlah', 'count'),
    ('Selamat', 'sum'),
    ('Tingkat_%', lambda x: (x.sum() / len(x)) * 100)
]).round(2)

print("\nTingkat keselamatan berdasarkan kelompok usia:")
print(analisis_usia)

print("\n--- Interpretasi Q6 ---")
print("""
Dari analisis berdasarkan kelompok usia, dapat dilihat bahwa:
- Anak-anak (usia 0-12 tahun) memiliki tingkat keselamatan tertinggi
- Dewasa muda (usia 19-35 tahun) memiliki tingkat keselamatan terendah
- Ini konsisten dengan kebijakan "women and children first" pada saat itu
- Faktor usia signifikan mempengaruhi peluang keselamatan di Titanic
""")

# Analisis korelasi gender
print("\n--- Analisis Tambahan: Pengaruh Gender ---")
analisis_gender = df.groupby('Sex')['Survived'].agg([
    ('Jumlah', 'count'),
    ('Selamat', 'sum'),
    ('Tingkat_%', lambda x: (x.sum() / len(x)) * 100)
]).round(2)

print(analisis_gender)
print("\nWanita memiliki peluang keselamatan yang jauh lebih tinggi")
print("dibanding pria (konsisten dengan kebijakan 'women and children first')")

# Statistik deskriptif lengkap
print("\n" + "=" * 60)
print("STATISTIK DESKRIPTIF DATASET TITANIC")
print("=" * 60)
print(df.describe().round(2))
```

Penjelasan:
- Gender paling signifikan (74.20% wanita dan 18.89% pria)
- Artinya wanita memiliki peluang keselamatan yang jauh lebih tinggi
dibanding pria (konsisten dengan kebijakan 'women and children first')
